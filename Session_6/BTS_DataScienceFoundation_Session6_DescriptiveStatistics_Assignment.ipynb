### Read data into a Dataframe:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('C:\BTS_Master\BTS_DataScienceFoundations/Session_6/FileData/1_titanic_dataset.csv')
```

```python
df.info()
```

## Basic statistics 

Sum, Cumulative sum, Count the number of non-NA values, Count the number of NA values, Minimum value of "Age" and "Passenger fare", Minimum value of "Age" and "Passenger fare", Range of "Age" and "Passenger fare" and Frequency table of "Age" and "Passenger fare".

```python
df.describe()[['Age', 'Fare']].T
```

```python
print('Age sum: ', df.Age.sum())
print('Passenger fare sum: ', df.Fare.sum())
print()
print('Age NA values: ', df.Age.isnull().sum())
print('Passenger fare NA values: ', df.Fare.isnull().sum())
print()
print('Age min: ', df.Age.min())
print('Passenger fare min: ', df.Fare.min())
print()
print('Age max: ', df.Age.max())
print('Passenger fare max: ', df.Fare.max())
print()
print('Age range: ', df.Age.max() - df.Age.min())
print('Passenger fare range: ', df.Fare.max() - df.Fare.min())
```

```python
df.Age.value_counts()
```

```python
df.Fare.value_counts()
```

```python
df.Age.cumsum()
```

```python
df.Fare.cumsum()
```

## Central Tendency

Mean and Median.

```python
print('Age mean: ', df.Age.mean())
print('Age median: ', df.Age.median())
print()
print('Passenger fare mean: ', df.Fare.mean())
print('Passenger fare median: ', df.Fare.median())
```

## Measures of spread

Range, Variance, Standard deviation, Q1, Q2, Q3, Box-and-whisker plots and Histogram.

```python
df.describe()[['Age', 'Fare']].T
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
print('Interquantile range: ', Q3 - Q1)
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
print('Interquantile range: ', Q3 - Q1)
```

```python
print('Age range: ', df.Age.max() - df.Age.min())
print('Passenger fare range: ', df.Fare.max() - df.Fare.min())
```

```python
df['Age'].var()
```

```python
df['Fare'].var()
```

```python
fig, axs = plt.subplots(1,2, figsize=[14,8])

df.boxplot('Age', ax=axs[0])
df.boxplot('Fare', ax=axs[1])
```

```python
plt.figure(figsize=[14,8])
df['Age'].hist(bins=20)
```

```python
plt.figure(figsize=[14,8])
df['Fare'].hist(bins=30)
```

## Finding outliers

```python
df['Age_stdv'] = abs(df['Age']-df['Age'].mean()) / df['Age'].std()
```

```python
df[df['Age_stdv'] > 2]
```

```python
df['Fare_stdv'] = abs(df['Fare']-df['Fare'].mean()) / df['Fare'].std()
```

```python
df[df['Fare_stdv'] > 2]
```

## Skewness and Kurtosis

```python
print('Age Skewness: ', df.Age.skew())
print('Age Kurtosis: ', df.Age.kurt())
print()
print('Passenger fare skewness: ', df.Fare.skew())
print('Passenger fare kurtosis: ', df.Fare.kurt())
```
