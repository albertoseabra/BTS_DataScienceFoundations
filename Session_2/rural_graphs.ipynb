```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
df = pd.read_csv('C:\BTS_Master\BTS_DataScienceFoundations\Session_1/file_formats/rural_population_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,8))
sns.barplot('UrbanRuralDesignation', 'Population2010', data=df)
```

#### the population is well divided between rural and urban with a big standard deviation

```python
plt.figure(figsize=(10,8))
sns.barplot('Continent', 'Population2010', data=df)
```

#### most of the population comes from Asia who also has the largest stdv

```python
plt.figure(figsize=(10,8))
sns.violinplot(y='Population2010', x='Continent', data=df)
```
